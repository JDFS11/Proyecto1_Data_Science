Preguntas

1)Cuales son las 10 compañias que mas producen peliculas, y cual es el ingreso promedio?

2)Que genero de peliculas es mas producido en la actualidad, y como se ha comportado la popularidad de estas peliculas con el paso del tiempo?

3)El presupuesto de las peliculas esta relacionado con la duracion de estas?

4)Que clasificacion de peliculas eran mas  producidas antes del año 2000?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Pregunta 1 

```python
Data = pd.read_csv('./movies.csv', sep=',')
Frequency = Data['company'].value_counts()
Frequency = Frequency.head(10)
New_Data = pd.DataFrame({'Compañias':Frequency.index, 'Cantidad de peliculas':Frequency})
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x='Compañias',y='Cantidad de peliculas', data=New_Data)
plt.axhline(y=400)
```

```python
DataNew = []
for i in Frequency.index:
  DataN = Data[Data['company'] == i]
  DataN.dropna()
  mean= DataN['gross'].mean()
  DataNew.append(mean)
DataF = pd.DataFrame({'Compañias':Frequency.index, 'Ingresos':DataNew})
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x='Compañias',y='Ingresos', data=DataF)
plt.axhline(y=200)
```

# Pregunta 2

```python
New_Data2 = Data[Data['year'] > 2017]
Frequency = New_Data2['genre'].value_counts()
Frequency_i = Frequency.index
plt.pie(Frequency, labels=Frequency_i,
        autopct='%1.1f%%', shadow=False, radius=3)
```

```python
Aux = []
year = np.arange(1980,2020)
for i in range(1980,2020):
  DataN = Data[Data['year'] == i]
  DataN = DataN[DataN['genre'] == 'Action']
  Value = DataN['genre'].size
  Aux.append(Value)
Df = pd.DataFrame({'Año':year, 'Cantidad de peliculas':Aux})
sns.lmplot(x='Año', y='Cantidad de peliculas', data=Df)
```

# Pregunta 3

```python
DataFra = pd.DataFrame({'Presupuesto':Data['budget'], 'Runtime':Data['runtime']})
DataFra = DataFra.dropna()
sns.lmplot(x='Runtime', y='Presupuesto', data=DataFra)
```

# Pregunta 4

```python
DataFram = pd.DataFrame({'Clasificacion':Data['rating'], 'Año':Data['year']})
DataFram = DataFram.dropna()
DataN2 = DataFram[DataFram['Año'] < 2001]
Frequencia = DataN2['Clasificacion'].value_counts()
DataFram = pd.DataFrame({'Clasificacion':Frequencia.index, 'Cantidad':Frequencia})
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x='Clasificacion',y='Cantidad', data=DataFram)
plt.axhline(y=2000)
```
